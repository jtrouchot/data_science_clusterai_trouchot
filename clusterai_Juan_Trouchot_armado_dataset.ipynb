# **Cargamos los datasets y nos quedamos con aquellos datos que tienen sexo, edad y provincia definida**


```
#defunciones = pd.concat([def2005, def2006, def2007, def2008, def2009, def2010, def2011, def2012, def2013, def2014, def2015, def2016, def2017], axis=0)
defunciones = pd.concat([def2010, def2011, def2012, def2013, def2014, def2015, def2016, def2017], axis=0)
sexo_max = 2
residencia_id_max = 95
edad_id_max = 17
defunciones = defunciones.loc[(defunciones.sexo_id <= sexo_max) & (defunciones.jurisdiccion_de_residencia_id <= residencia_id_max) & (defunciones.grupos_edad_id <= edad_id_max)] #nos quedamos solo con los sexos varon o mujer y con las provincias especificadas
defunciones.head(15)
```

# **Aquí se agregará la feature "poblacion" al dataset defunciones**

```
defunciones['poblacion'] = 0

for x in range(0,defunciones.shape[0]):
  for y in range(4,proy_pobl.shape[1]-8):
    if (defunciones['año'].values[x] == proy_pobl.columns.values[y]):
      for z in range(0,proy_pobl.shape[0]):
        if (defunciones['jurisdiccion_de_residencia_id'].values[x] == proy_pobl['codigo_provincia'].values[z] and defunciones['sexo_id'].values[x] == proy_pobl['codigo_genero'].values[z]):
          defunciones['poblacion'].values[x] = proy_pobl.values[z][y]
```

# **Agregamos la cantidad de defunciones per capita al dataframe**

```
defunciones['cantidad_de_defunciones_percapita'] = 0
defunciones.cantidad_de_defunciones_percapita = defunciones.cantidad_de_defunciones / defunciones.poblacion
defunciones.head()
```
